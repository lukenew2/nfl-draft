```python
import pandas as pd
```

```python
roster = pd.read_html('http://www.cfbstats.com/2016/team/811/roster.html')
```

```python
roster2018 = roster[0]
```

```python
punting = pd.read_html('http://www.cfbstats.com/2016/team/811/punting/index.html')
ko = pd.read_html('http://www.cfbstats.com/2016/team/811/kickoff/index.html')
kick = pd.read_html('http://www.cfbstats.com/2016/team/811/kicking/index.html')
score = pd.read_html('http://www.cfbstats.com/2016/team/811/scoring/index.html')
to = pd.read_html('http://www.cfbstats.com/2016/team/811/total/index.html')
allpurpose = pd.read_html('http://www.cfbstats.com/2016/team/811/allpurpose/index.html')
interception = pd.read_html('http://www.cfbstats.com/2016/team/811/interception/index.html')
tackle = pd.read_html('http://www.cfbstats.com/2016/team/811/tackle/index.html')
sack = pd.read_html('http://www.cfbstats.com/2016/team/811/sack/index.html')
passing = pd.read_html('http://www.cfbstats.com/2016/team/811/passing/index.html')
```

```python
punting = punting[0]
kickoffs = ko[0]
place_kicking = kick[0]
scoring = score[0]
total_offense = to[0]
all_purpose_running = allpurpose[0]
interceptions = interception[0]
tackles = tackle[0]
sacks = sack[0]
passing = passing[0]
```

```python
punting.dropna(inplace=True)
kickoffs.dropna(inplace=True)
place_kicking.dropna(inplace=True)
scoring.dropna(inplace=True)
total_offense.dropna(inplace=True)
all_purpose_running.dropna(inplace=True)
interceptions.dropna(inplace=True)
tackles.dropna(inplace=True)
sacks.dropna(inplace=True)
passing.dropna(inplace=True)
```

```python
roster2018.head()
```

```python
def clean_names(df_column):
    
    new_column = []

    for name in df_column:
        
        names = name.split(',')
        new_column.append(names[1].strip() + ' ' + names[0])
        
    rosternames = pd.Series(new_column)
    roster2018['Name'] = rosternames
clean_names(roster2018['Name'])
```

```python
kickoffs = kickoffs[['Name', 'Kickoffs', 'Yards', 'Touchback']]
passing = passing[['Name', 'Att', 'Comp', 'TD', 'Int']]
passing.rename(columns={'TD': 'Passing TD', 'Int': 'Passing Int'}, inplace=True)
punting = punting[['Name', 'Punts', 'Yards']]
scoring = scoring[['Name', 'G', 'TD', 'FG', '1XP', '2XP', 'Safety', 'Points']]
total_offense = total_offense[['Name', 'Rush Yards', 'Pass Yards', 'Plays', 'Total Yards', 'Yards/Play']]
all_purpose_running = all_purpose_running[['Name', 'Recv.', 'Punt Ret.', 'Kick Ret.', 'Total Yards']]
interceptions = interceptions[['Name', 'Int.']]
tackles = tackles[['Name', 'Solo', 'Assisted']]
sacks = sacks[['Name', 'Sacks', 'Sack Yards']]
```

```python
wy1 = roster2018.merge(kickoffs, how='outer',on='Name')
wy0 = wy1.merge(passing, how='outer',on='Name')
wy2 = wy0.merge(punting, how='outer',on='Name')
wy3 = wy2.merge(scoring, how='outer',on='Name')
wy4 = wy3.merge(total_offense, how='outer',on='Name')
wy5 = wy4.merge(all_purpose_running, how='outer',on='Name')
wy6 = wy5.merge(interceptions, how='outer',on='Name')
wy7 = wy6.merge(tackles, how='outer',on='Name')
wy8 = wy7.merge(sacks, how='outer',on='Name')
```
